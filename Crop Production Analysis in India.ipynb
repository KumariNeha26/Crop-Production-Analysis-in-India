```python
# Clean the data by removing rows with null values
data = pd.read_csv(r"C:\Users\kushn\Downloads\Crop Production data.csv")  

# Clean the data by removing rows with null values
data_cleaned = data.dropna()
```

```python
# Visualization: Crop Production by Crop
plt.figure(figsize=(20, 5))
sns.barplot(data=data_cleaned, x='Crop', y='Production', estimator=sum,color='red')
plt.title('Crop Production by Crop')
plt.xlabel('Crop')
plt.ylabel('Total Production')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization: Distribution of Production
plt.figure(figsize=(12, 5))
sns.histplot(data_cleaned['Production'], bins=30, kde=True, color='yellow')
plt.title('Distribution of Crop Production')
plt.xlabel('Production')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization: Crop Production by State
plt.figure(figsize=(20, 5))
sns.barplot(data=data_cleaned, x='State_Name', y='Production', estimator=sum, color='green')
plt.title('Crop Production by State')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization: Crop Production Trends Over Years
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20, 6))
sns.lineplot(data=data_cleaned, x='Crop_Year', y='Production', estimator='sum',color='pink')
plt.title('Crop Production Over the Years')
plt.xlabel('Year')
plt.ylabel('Production')
plt.show()
```

```python
# Visualization: Crop Production by Season
plt.figure(figsize=(20, 6))
sns.barplot(data=data_cleaned, x='Season', y='Production', estimator=sum, color='orange')
plt.title('Crop Production by Season')
plt.xlabel('Season')
plt.ylabel('Total Production')
plt.show()
```

```python
# Correlation heatmap
numeric_features = data_cleaned.select_dtypes(include=[np.number])
plt.figure(figsize=(6, 4))
sns.heatmap(numeric_features.corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```
